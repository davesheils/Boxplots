## Summarise the history of the box plot and situations in which it used.

**Description**

In descriptive statistics, a boxplot or box and whiskers plot is a method for visualising groups of numerical data through their quartiles. (wikipedia). The box plot consists of a rectangular box (representing the interquartile range of values) and two "whiskers", i.e. lines extending to indicate the range of values below or above the interquartile range (which may or may not include values considered to be outliers).The box itself is divided by a line indicating the median.

Although the basic form of the boxplot was established in the early 1950s, the boxplot as used today was developed and made more widley known by John Tukey in a paper published in 1977. (OU, 2010; Stryjewski & Wickham, 2011).

Box plots "use robust summary statistics that are always located at actual data points, are quickly computable (originally by hand) [...] They are particularly useful for comparing distributions across groups" (Stryjewski & Wickham, 2011).

One big advantage of boxplots is that they can readily demonstrate the extent to which a dataset is symetrical or skewed one way or the other. In addition, they can be used to identify outliers and show what their values are. how tightly your data is grouped. (Galarnyk, 2018)


The simplest form of a boxplot can be drawn manually as follows:

1. Draw the axis, mark the scale and add an axis label. Add a title and the source of the data.
2. Mark the five key summary values (minimum, Q1, median, Q3 and Maximum) with vertical lines
3. Add the box, hinges and whiskers  and label the vertical lines with the five key summary values.

(OU, 2010, Book D, p12)

When calculating the following datasets manually, it is usual to write out the values in increasing order. However, this is obviously impractical for larger datasets. Therefore the next sections will demostrate the use of the library matplotlib.pyplot to draw boxplots.


The following code will create a simple boxplot based on numbers randomly generated by np.random from the a normal distribution.

```python
# Create a simple boxplot

import numpy as np
import matplotlib.pyplot as plt

import pandas as pd



# create a boxplot from the sepal length column in the dataset. by default, pyplot boxplot function produces
# a vertical boxplot. In the version below, we are producing a horizonal boxplot by setting the parameter vert to false.

x = np.random.normal(0, 1, 1000)
plt.rcParams['figure.figsize'] = [10, 8] # to set size of the output of pyplot
plt.boxplot(x, vert = False)

plt.title("Example 1: Simple boxplot (with outliers)")
plt.xlabel("Randomly generated values")

plt.show()
```

The whiskers can indicate either the maximum or minimum values of the dataset. More usually, the whiskers will indicate that the lowest datum still within 1.5 IQR of the lower quartile, and the highest datum still within 1.5 IQR of the upper quartile (often called the Tukey boxplot). This is the default setting for the whiskers in the pybolt.boxplot function. Other variations may be to set the whiskers at the 9th and 91st percentile (wikipedia). In these implementations, as in Example 1 points beyond the reach of the whiskers are considered outliers and may be plotted seperately. ## Demonstrating the use of the boxplot using data

The following pair of boxpolots show how they can be used to compare the spread of two related datasets. The datasets are the respective salaries of San Francisco city  male and female employees.

This dataset covers the years 2011 to 2014. In addition, the original dataset contains names of employees. For this exercise, I have modified the dataset, removing names. The original dataset contained salaries of part-time employees, but for comparibility, they have been removed from this analysis. 


The original dataset may be accessed here at [kaggle](https://www.kaggle.com/ronaldtroncoso20/sf-salaries-extended) or from [Ronald Trosconos github page](https://github.com/RonaldTroncoso/SF_Salary_Trends).


```python
df = pd.read_csv('SF_salary_data_gender.csv')

male = df[df['gender'] == 'male'] # select all rows from species where gender of employee == male
female = df[df['gender'] == 'female'] # select all rows from species where gender of employee == female

```

```python
# Now create two boxplots

plt.figure(figsize=(16, 20)) # This sets the size of the individual plots 

plt.subplot(2,1,1)
plt.boxplot(male['BasePay'], vert = False) 
plt.xlabel("Base pay of Full-time Male Employees")
plt.title("Distribution of Male Employees Salaries - SF City")

plt.subplot(2,1,2)
plt.boxplot(female['BasePay'], vert = False) 
plt.xlabel("Base pay of Full-time Female Employees")
plt.title("Distribution of Female Employees Salaries - SF City")


plt.show()
```

Boxplots can also be drawn on the same plot for further ease of comparison.

```python
data = [male['BasePay'],female['BasePay']]
plt.boxplot(data)
plt.title('Distributions of Male and Female Salaries compared (Male = 1, Female = 2)')
plt.show()
```

In the above examples, we can see that when the outliers are excluded, male and female salaries lie roughly in the same range. However, the median salary of female employess is less than that of males.

### Violin Plots

One limitation sometimes noted with boxplots (e.g. by K. Potter) is the inability of boxplots to tell us anything about modalities, i.e the most comonly occuring values. Therefore similarly looking boxplots may conceal the varying modalities of different datasets. Therefore a number of variations of the boxplot have evolved such as the histplot, the beanplot and the violin plot.


```python
plt.figure(figsize=(10, 10))

plt.subplot(2,1,1)
plt.hist(male['BasePay']) 
plt.xlabel("Base pay of Full-time Male Employees")
plt.title("Hist1 - Distribution of Male Employees Salaries - SF City")

plt.subplot(2,1,2)
plt.hist(female['BasePay']) 
plt.xlabel("Base pay of Full-time Female Employees")
plt.title("Hist2 - Distribution of Female Employees Salaries - SF City")
```

## Explanation of terminogy

The data is displayed based on a five number summary:

**Minimum** - This is the lowest value in the dataset.

**Maximum** This is the highest value in the dataset.

**Median** - The median is the middle value in a (sorted) data series. For for an data series with an even number of values, 
the median is the average of the two middle values. The median can also be referred to as the **2nd quartile**.

**Interquartile range** - This is the middle 50% of values.  As a measure of spread, the interquartile range is calculated by subtracting the 3rd Quartile for from the first **(Q3 - Q1)**. If the dataset is skewed, the IQR would be considered a better measure of spread that the standard deviation.

**1st quartile (Q1)** - This is the lower point of the interquartile range. It is the median of the lower half of the dataset.

**3rd quartile (Q3)** This is the upper point of the interquartile range and is the median of the upper half of the dataset.

**Percentile ** This is a measure used in statistics indicating the value below which a given percentage of observations in a dataset falls. For example, the 20th percentile is the value below which 20% of the observations may be found. 

**Outliers** In implementations where the whiskers do not represent the maximum and minimum values of the dataset, outliers are values that lie beyond the range between the two whiskers and are plotted seperately.
