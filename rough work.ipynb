```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 7.5] # to set the default size of the output of pyplot
import pandas as pd
```

```python
SFSalary = pd.read_csv('SF_salary_data_gender.csv') 
x = SFSalary['BasePay']
x.head()
```

```python
plt.boxplot(x, vert = False)
plt.show()
```

```python
SalaryM = SFSalary[SFSalary['gender'] == 'm']
SalaryF = SFSalary[SFSalary['gender'] == 'f']
```

```python
plt.figure(figsize=(16, 20)) # This sets the size of the individual plots 

plt.subplot(2,1,1)
plt.boxplot(SalaryM['BasePay'], vert = False) 
plt.xlabel("Base Pay, 2011 (USD)")
plt.title("Boxplot Mens Salary, SF Municipality 2011")

plt.subplot(2,1,2)
plt.boxplot(SalaryF['BasePay'], vert = False)
plt.xlabel("Base Pay, 2011 (USD)")
plt.title("Boxplot Womens Salary, SF Municipality 2011")

plt.show()
```
